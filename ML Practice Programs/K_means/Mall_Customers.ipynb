```python
import matplotlib.pyplot as plt 

plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'])


plt.legend()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3)
km
```

```python
clustor = km.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
clustor
```

```python
km.cluster_centers_
```

```python
df['Clustor'] = clustor
```

```python
df.head()
```

```python
df1 = df[df['Clustor']==0]
df2 = df[df['Clustor']==1]
df3 = df[df['Clustor']==2]
```

```python
import matplotlib.pyplot as plt 

plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color="red",label='red')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color="green",label='green')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color="black",label='black')
plt.scatter(km.cluster_centers_[: ,0],km.cluster_centers_[: ,1],color="yellow",marker="*")

plt.legend()
plt.show()
```

```python
sse = []
k_range = range(1,11)

for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_range,sse,label="Ellobow Line")
plt.xlabel('k')
plt.ylabel('Sum of Sqaured Method')
plt.legend()
```
