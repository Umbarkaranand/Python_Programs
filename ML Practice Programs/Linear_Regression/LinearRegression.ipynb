```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("LinearRegression.csv")
print(df)
```

```python
import matplotlib.pyplot as plt
x = df['X']
y = df['Y']
plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=23)
```

```python
x_train = np.array(x_train).reshape(-1,1)
x_train
```

```python
from sklearn.linear_model import LinearRegression
rg = LinearRegression()

```

```python
rg.fit(x_train,y_train)
```

```python
y_predict_train = rg.predict(x_train)
y_predict_train
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,y_predict_train,color="red")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
from sklearn.metrics import r2_score
r2_score(x_train,y_predict_train)
```

```python
x_test = np.array(x_test).reshape(-1,1)
x_test
```

```python
y_predict_test = rg.predict(x_test)
y_predict_test
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict_test,color="red")
plt.xlabel("X")
plt.ylabel("Y")
```
