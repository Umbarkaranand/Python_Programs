```python
import seaborn as se
import matplotlib.pyplot as plt
```

```python
def cat_count(col):
    plt.figure(figsize=(8,5))
    plt.subplot(1,2,1)
    se.countplot(x=df[col])
    plt.legend()
```

```python
cat_count('Survived')
plt.title("totla No. of Survied")
```

```python
def cat_pie(col):
    plt.subplot(1,2,2)
    df1[col].value_counts().plot(kind='pie',autopct="%0.2f")
    plt.legend()
```

```python
cat_count('Survived')
plt.title("totla No. of Survied")

cat_pie('Survived')
plt.title("totla No. of Survied")
```

```python
cat_count('Sex')
plt.title("totla No. of Survied")

cat_pie('Sex')
plt.title("totla No. of Survied")
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1,2,1)
se.histplot(x=df['Age'],bins=6)
plt.title("Distribution of Age histograma")
plt.legend()

plt.subplot(1,2,2)
se.kdeplot(x=df['Age'])
plt.title("Distribution of Age Kdeplot")
plt.legend()

```

```python
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
se.histplot(x=df['Fare'],bins=4)
plt.title('Disribution of Fare')
plt.legend()

plt.subplot(1,2,2)
se.kdeplot(x=df['Fare'])
plt.title('Distribution of Fare Kdeplot')
plt.legend()
```

```python
flit = df['Age'] < 10
df.loc[flit]['Survived'].value_counts().plot(kind='pie',autopct="%0.2f")
plt.title('if you are below the age of 10 u have 60.29 percentage chances of survival')
```

```python
flit = df['Fare'] > 50
df.loc[flit]["Survived"].value_counts().plot(kind='pie',autopct="%0.2f")
plt.title('if you are below the age of 10 u have 68.29 percentage chances of survival')
```
